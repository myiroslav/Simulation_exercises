```python
## Numerical application
#K = 105, T = 1, r = 0.001, sigma = 0.03, S_0 = 100
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import norm

def strike( T = 1, r= 0.001, sigma = 0.03, S0 = 100):
  B = np.random.normal(0, np.sqrt(T))
  return S0 * np.exp( (r - sigma**2/2)*T + sigma* B)
  
def strike_path(n, T = 1, r= 0.001, sigma = 0.03, S0 = 100) : # simulate n samples of S_T
    S = np.zeros(n +1)
    h = T/n
    S[0] = S0
    for i in range (1,n+1):
        B = np.random.normal(0, np.sqrt(h), n)
        S[i] = S[i-1]+ r*S[i-1] *h + sigma * S[i-1]* B[i-1]
    return S
    
```

```python
def plot_strikes(n, k = 5, T = 1, r= 0.001, sigma = 0.03, S0 = 100) :
    Time = np.linspace(0, T, n+1)
    plt.figure()
    for i in range(k):
        S = strike_path(n, T, r, sigma , S0)
        plt.plot(Time, S)
    plt.show()

plot_strikes(1000)
```

```python
#awfull complexity for expectations and probabilities
def compute_expectation_path(T = 1, n_times = 100, n_samples = 1000, K = 105,r = 0.001, sigma = 0.03, S0 = 100 ):
    E = 0
    P = 0
    for i in range(n_samples):
        S = strike(n_times, T, r, sigma, S0)
        if S[-1] > K:
            E += S[-1]
            P +=1
    return E/n_samples, P/n_samples
#better compexity
def compute_expectation_distrib(T = 1, n_samples = 1000, K = 105,r = 0.001, sigma = 0.03, S0 = 100 ):
    E = 0
    P = 0
    for i in range(n_samples):
        B = np.random.normal(0, np.sqrt(T))
        S_T = S0 * np.exp( (r - sigma**2/2)*T + sigma* B)
        if S_T > K:
            E += S_T
            P+=1
    return E/n_samples, P/n_samples

def call_value(K,T,r, sigma, S0 ):
    d2 = 1/(sigma*np.sqrt(T))*(np.log(S0/K)+(r-sigma**2/2)*T)
    d1 = d2 + sigma*np.sqrt(T)
    phi1 = norm.cdf(d1)
    phi2 = norm.cdf(d2)
    return S0 * phi1 - K * np.exp(-r*T)*phi2
```

```python
def plot_expect (n_step = 100, T = 1, n_samples = 1000, K_min = 95, K_max = 105,r = 0.001, sigma = 0.03, S0 = 100 ):
    simulated_expect_val = np.zeros(n_step)
    theoretic_expect_val = np.zeros(n_step)
    theoretic_proba_val = np.zeros(n_step)
    simulated_proba_val = np.zeros(n_step)
    K_val = np.linspace(K_min, K_max, n_step)
    for K,i in zip (K_val, range(n_step)):
        simulated_expect_val[i], simulated_proba_val[i] = compute_expectation_distrib(T, n_samples, K, r, sigma, S0)
        d2 = 1/(sigma*np.sqrt(T))*(np.log(S0/K)+(r-sigma**2/2)*T)
        d1 = d2 + sigma*np.sqrt(T)
        theoretic_expect_val[i] = np.exp(r*T)* S0*norm.cdf(d2)
        theoretic_proba_val[i] = norm.cdf(d1)
    fig, ax = plt.subplots(1)
    ax.plot(K_val, simulated_expect_val, label = 'Simulated expectation of S_T > K')
    ax.plot(K_val, theoretic_expect_val, label = 'Theoretic expectation of S_T > K')
    ax.set_ylabel('Expected payoff, depending on K')
    ax.set_xlabel(' K ')
    ax.legend()
    ax.set_title('Expectation of payoff and probability of strictly positive payoff, depending on K')

    fig, ax = plt.subplots(1)
    ax.plot(K_val, simulated_proba_val, label = 'Simulated proba of S_T > K')
    ax.plot(K_val, theoretic_proba_val, label = 'Theoretic proba of S_T > K')
    ax.set_ylabel('Probability of strictly positive payoff')
    ax.set_xlabel(' K ')
    ax.legend()
    
    
    plt.show()
        
        
def plot_call(T = 1, n_step = 1000, n_samples = 1000 ,K_min = 95, K_max = 115,r = 0.001, sigma = 0.03, S0 = 100, path =False):
    K_val = np.linspace(K_min, K_max, n_step)
    simulated_expect_val= np.zeros(n_step)
    simulated_proba_val = np.zeros(n_step)
    call_simulated = np.zeros(n_step)
    C_val = np.zeros(n_step)
    
    for (K,i) in zip(K_val, range(n_step)):
        simulated_expect_val[i], simulated_proba_val[i] = compute_expectation_distrib(T, n_samples, K, r, sigma, S0)
        #Using formula from question 4)
        C_val[i] = call_value(K, T, r, sigma, S0)
    #Using formula from question 1a)    
    call_simulated = np.exp(-r*T)* simulated_expect_val - np.exp(-r*T) * K_val * simulated_proba_val
    
    fig, ax = plt.subplots(1)
    ax.plot(K_val, C_val, label = 'Calculated values of Call depending on K')
    ax.plot(K_val, call_simulated, label = 'Call with simulated only S_T')
    ax.set_xlabel('K')
    ax.set_ylabel('Call value depending on K')
    ax.legend()
    ax.set_title('Call value depending on K')
    plt.show()
```
